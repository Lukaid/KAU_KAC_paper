```python
headers = {'User-Agent': 'Mozilla/5.0'} 

error_url = []

for i in range(33):
    covid_news_list = []
    start = day_list_1912_2010[i*2]
    end = day_list_1912_2010[i*2+1]

    df = pd.read_csv(f'news_url_{start}_{end}.csv')

    news_list = []
    for i in range(len(df)):
        try:
            news_dict = {}
            url = df['0'][i]

            req = requests.get(url, headers = headers)
            soup = BeautifulSoup(req.text, 'html.parser')

            table = soup.find('div',{'id' : 'main_content'})
            title = table.find('h3', {'id':'articleTitle'})
            body = table.find('div', {'id':'articleBodyContents'})

            news_dict['title'] = title.get_text()
            news_dict['body'] = body.get_text()
            news_list.append(news_dict)
        except:
            print(f'오류 발생, {start}_{end}의 {i}번째')
            error_url.append(url)
            pass

    article_list = pd.DataFrame(news_list)
    article_list.to_csv(f'article_list_{start}_{end}.csv', encoding='utf-8-sig')
    print(f'article_list_{start}_{end}: 완료')
    print(f'기사 개수: {len(news_list)}')
```
